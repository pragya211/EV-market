# Analysis

```python
# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10), sharex=True, sharey=True)

# Plot data for each DataFrame
axs[0, 0].plot(e2w['Year'], e2w['Total'], marker='o', color='blue')
axs[0, 0].set_title('Electric 2 wheeler')
axs[0, 0].set_ylabel('Total Sales')

axs[0, 1].plot(e3w['Year'], e3w['Total'], marker='o', color='green')
axs[0, 1].set_title('Electric 3 wheeler')

axs[1, 0].plot(e4w['Year'], e4w['Total'], marker='o', color='red')
axs[1, 0].set_title('Electric 4 wheeler')
axs[1, 0].set_xlabel('Fiscal Year')
axs[1, 0].set_ylabel('Total Sales')


# Adjust layout to avoid overlap
plt.tight_layout()

plt.show()
```

```python
# Function to calculate growth rate
def calculate_growth_rate(df):
    df['Growth Rate'] = df['Total'].pct_change() * 100
    return df

    
# Calculate growth rate for each DataFrame
e2w = calculate_growth_rate(e2w)
e3w = calculate_growth_rate(e3w)
e4w = calculate_growth_rate(e4w)
   
# Combine results into a single DataFrame
e2w['Source'] = 'Electric 2 wheeler'
e3w['Source'] = 'Electric 3 wheeler'
e4w['Source'] = 'Electric 4 wheeler'


combined_df = pd.concat([e2w, e3w, e4w])

# Plotting
plt.figure(figsize=(12, 8))
for label, df_group in combined_df.groupby('Source'):
    print(f"Growth Rates for {label}:")
    print(df_group[['Year', 'Growth Rate']])
    print()
    plt.plot(df_group['Year'], df_group['Growth Rate'], marker='o', linestyle='-', label=label)

plt.title('Growth Rate Over Years for Different DataFrames')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.legend()
plt.grid(True)
plt.show()
```
